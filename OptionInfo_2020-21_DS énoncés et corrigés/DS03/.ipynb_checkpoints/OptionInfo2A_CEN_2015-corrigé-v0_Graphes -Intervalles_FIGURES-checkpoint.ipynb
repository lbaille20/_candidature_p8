```python
## Analyse, le pas est (1 - (-1)) / 2 * 2 ** p
## s_i = -1 + i * pas = -1 + i * (1 - (-1)) / 2 * 2 ** p = -1 + i * 1 / 2 ** p
## la position du noeud d'index 2**p + k (0 <= k < 2 ** p - 1) [noeud d'indice 0 dans le niveau p] est 
## -1 + (2 * k + 1) * 1 / 2 ** p

## calcul du couple (p, k)  partir de l'indice
## i = 2**p + k
## rciproquement 
## 2**p <= i < 2**p + 2**p
## 2**p <= i < 2**(p + 1)
## p <= log_2(i) < p + 1
## p = floor( log_2(i) )
## k = i - 2**p
```

```python
## fonction de calcul
def canonical_positions_binary_tree(i):
    from math import floor, log2
    p = floor( log2(i)); k = i - 2**p
    return (-1 + (2 * k + 1) * 1 / 2 ** p, -p)
```

```python
def complete_binary_tree(h):
    import networkx as nx
    G = nx.Graph()
    ## binary tree
    ## height = h ; nb total de noeuds 2**h
    h = 6
    G.add_node(1)
    for p in range(1, h):
        for k in range(2**p):
            i_fils = 2**p + k
            G.add_node(i_fils) 
            G.add_edge(i_fils // 2, i_fils)
    return G

def complete_binary_tree_positions(h):
    pos = dict()
    for p in range(h):
        for k in range(2**p):
            s = np.linspace(-1, 1, 2*(2**p) + 1)
            #pos[2**p + k] = (2**p*(k - 2**p / 2), -p)
            pos[2**p + k] = (s[2*k + 1], -p)
    return pos

def draw_complete_binary_tree(nodes, edges, positions):
    a=nx.draw(G, pos=positions, with_labels = True,
            node_size = 500, node_color = 'w', node_shape = 'o', edgecolors='k', alpha = 1)
    ## edgecolors : https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.scatter.html
    return a
```

```python
class Tree:
    def __init__(self, d = None, l = None, r = None):
        self.left = l
        self.right = r
        self.data = d
    def add_data(self, d):
        self.data = d
    def add (self, l, r, d = None):
        self.left = l
        self.right = r
        self.data = d
    def add_child(self, c, side = 'r', data = None):
        if side == 'l':
            self.left = c
        else:
            self.right = c
    def is_leaf(self):
        return self.left is None and self.right is None
    def copy(self):
        if self.is_leaf():
            return Tree(self.data, None, None)
        return Tree(self.data, self.left.copy(), self.right.copy())
```

```python
def traverse_tree(t):
    #
    def aux(t, i):
        hl, hr, nodes_dic[i] = 0, 0, t.data
        if t.left is not None:
            edges_dic[(i, 2 * i)] = ''
            hl = aux(t.left, 2 * i)
        if t.right is not None:
            edges_dic[(i, 2 * i + 1)] = ''
            hr = aux(t.right, 2 * i + 1)
        return 1 + max(hl, hr)
    #
    nodes_dic , edges_dic = dict(), dict()
    h = aux(t, 1)
    return h, nodes_dic , edges_dic
```

```python
def draw_positional_binary_tree(nodes, edges, positions,
                               nodes_edges_color = None, nodes_color = None, nodes_labels = None,
                               edges_labels = None, padx = 0.3, pady = 0.3, z = 1.):
        G = nx.Graph()
        G.add_nodes_from(nodes)
        G.add_edges_from(edges)
        ##https://networkx.github.io/documentation/latest/reference/generated/networkx.drawing.nx_pylab.draw_networkx_nodes.html#networkx.drawing.nx_pylab.draw_networkx_nodes
        a = nx.draw_networkx_nodes(G, pos=positions, with_labels = True, nodelist = nodes,
                                   node_size = 500, node_color = nodes_color, node_shape = 'o',
                                   edgecolors = nodes_edges_color, alpha = 1)
        ## edgecolors : https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.scatter.html
        ## https://networkx.github.io/documentation/latest/reference/generated/networkx.drawing.nx_pylab.draw_networkx_edges.html#networkx.drawing.nx_pylab.draw_networkx_edges
        b = nx.draw_networkx_edges(GC, pos=positions, edgelist = edges)
        ## https://networkx.github.io/documentation/latest/reference/generated/networkx.drawing.nx_pylab.draw_networkx_labels.html#networkx.drawing.nx_pylab.draw_networkx_labels
        nx.draw_networkx_labels(G, pos=positions, labels = nodes_labels)
        ##
        X, Y = [x for x, y in positions.values()], [y for x, y in positions.values()]
        xmin, xmax, ymin, ymax = min(X), max(X), min(Y), max(Y)
        plt.axis([z * (xmin - padx), z * (xmax + padx), ymin - pady, ymax + pady])
        return a, b
```

### Matrice d'adjacence

La matrice d'adjacence d'un graphe indique si les sommets sont connects entre ou pas, par des artes.  
Et aussi, le cas chant, les poids attachs aux diffrentes artes, comme ici les distances entre les sommets. 

Si le graphe est orient, chaque arte comporte une flche qui indique si l'on va du sommet a au sommet b ou le contraire. 
On parle alors d'arc, plutt que d'arte.

On convient que l'on crira -1 lorsque deux sommets ne sont pas connects.  

Si le graphe n'est pas orient, la matrice d'adjacence est symtrique. 

S'il n'y pas de boucle (c'est--dire pas d'arte partant d'un sommet et revenant sur ce mme sommet), comme dans ce sujet, on crira -1  l'intersection  de la ligne et de colonne correspondant  ce sommet.


Ici, chaque sommet est repr, par sa position dans la liste S :
- A position 0
- B position 1
...
- E position 4

### Construction de l'arbre d'exploration  
On parcourt la liste p.  
Pour chaque position dans p, on relie le sommet correspondant (dans S)  son fils (tiquet par le nom du pre),  
ce faisant, on construit la liste d'adjacence de ce sommet (ensemble de ses fils).

```python
def arborescence(S, p, e0):
    LAdj_bfs = []
    for i in range(len(p)):
        LAdj_bfs.append([])
        ## recherche des fils
        for j in range(len(p)):
            if p[j] == S[i]:
                LAdj_bfs[i].append((S[j], 1))
    return LAdj_bfs
```

```python
S = ['A', 'B', 'C', 'D', 'E']
p = [-1, 'A', 'B', 'E', 'A']
arborescence(S, p, e0 = 'A')
```

```python
def arborescence_complete(S, LAdj, p, e0):
    LAdj_bfs = arborescence(S, p, e0)
    LAdj_bfs2 = []
    for i in range(len(LAdj)):
        LAdj_bfs2.append([])
        for t in LAdj[i]:
            s, d = t
            if (s, 1) not in LAdj_bfs[i]:
                LAdj_bfs2[i].append((s, 1))
    return LAdj_bfs, LAdj_bfs2
```

## Algorithme de Dijkstra

```python
def dijkstra(LAdj, S, s0):
    n = len(LAdj) #nombre de sommets du graphe
    ## la fonction extraire_min(E) renvoie la position du premier lment ralisant le minimum dans une liste non vide E
    ## et le supprime de E
    def extraire_min(E, e):
        pos_min, val_min = 0, e[E[0]]
        for k in range(1, len(E)):
            if e[E[k]] < val_min:
                pos_min, val_min = k, e[E[k]]
        return E.pop(pos_min)
    ##
    e = n * [float('inf')] #initialisation des distances  s0   +  
    p = n * [-1] #initialisation des couleurs   blanc 
    E = list(range(n))
    e[S.index(s0)] = 0
    while not E == []:  ## tant que tous les sommets n'ont pas tous t traits (i.E. qu'une distance minimale n'a
                        ## pas t trouve pour accder  ce sommet depuis s0)
        print(E)
        u = S[extraire_min(E, e)]
        print('min :', u)
        for t in LAdj[S.index(u)]: #
            v, V = t
            if e[S.index(u)] + V < e[S.index(v)]:
                e[S.index(v)] = e[S.index(u)] + V
                p[S.index(v)] = u
        for s in S:
            if S.index(s) in E:
                print(s, (e[S.index(s)], p[S.index(s)]), end=' ; ')
            else:
                print(s, ("*", ' '), end=' ; ')
        print()
    return e, p
```

### Reconstruction du plus court chemin  partir des listes e et p (renvoyes)

```python
def plus_court_chemin(S, e, p, e0, s0):
    chemin = [s0]
    index_pere = S.index(s0)
    while S[index_pere] != e0:
        pere = p[index_pere]
        chemin.append(pere)
        index_pere = S.index(pere)
    return chemin[::-1]
```
